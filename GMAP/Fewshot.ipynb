```python
# Prompt- few shot

prompt = f"""
Bạn là hệ thống chuẩn hoá địa chỉ hành chính Việt Nam.
Hãy luôn trả JSON theo schema cho sẵn + thêm các trường dữ liệu vs các field như sau :   
- Province_ner: phần text nào trong raw tương ứng với tỉnh,
- district_ner: phần text tương ứng với huyện/quận,
- ward_ner: phần tương ứng với phường/xã/tt/phường thị trấn nếu có,
Và đồng thời chuẩn hóa (normalized) tên : Province_nor, District_nor,ward_nor nếu có (ví dụ chữ hoa, bỏ dấu, cách viết chuẩn,…)

Ví dụ 1:  
Input: "bệnh viện đa khoa huyện phù yên , 37 tt phù yên , tt phù yên , phù yên"  
Output: {{
  "Province_ner": "hà nội",
  "district_ner": "phù yên",
  "ward_ner": "tt phù yên",
  "Province_normalized": "Hà Nội",
  "district_normalized": "Huyện Phù Yên",
  "ward_normalized": "Thị trấn Phù Yên"
}}

Ví dụ 2:  
Input: "UBND phường Hoàng Liệt , quận Hoàng Mai , Hà Nội"  
Output: {{
  "Province_ner": "Hà Nội",
  "district_ner": "quận Hoàng Mai",
  "ward_ner": "phường Hoàng Liệt",
  "Province_normalized": "Hà Nội",
  "district_normalized": "Quận Hoàng Mai",
  "ward_normalized": "Phường Hoàng Liệt"
}}

Giờ đến input là : "VTV- 8 Pham hung, Nam Từ Liêm, Hà Nội"  
Output:
"""
```
